```python
import geopandas as gpd

# Cargar el GeoDataFrame desde el archivo GeoJSON
gdf_barrios = gpd.read_file("barrios.geojson")
# Filtrar el GeoDataFrame para obtener solo los barrios de la localidad "Usme"
gdf_usme = gdf_barrios[gdf_barrios["localidad"] == "Usme"]
# Plotear los barrios de la localidad "Usme"
ax=gdf_usme.plot(
    column="barriocomu",
    legend=True,
    edgecolor='black',
    cmap='Dark2',
    legend_kwds={"loc":"upper left", "fontsize": 10 },
    figsize=(15, 20)
    )
# Agregar etiquetas (nombres de los barrios) a los puntos medios de los barrios
```
