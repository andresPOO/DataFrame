```python
import geopandas as gpd
gdf_localidades=gpd.read_file("localidades.geojson")
gdf_localidades
```

```python
gdf_localidades.plot("SHAPE_Length", figsize=(12,20))
```

```python
gdf_localidades["area"]=gdf_localidades.area
gdf_localidades
```

```python
gdf_localidades["Perimetro"]=gdf_localidades.boundary
gdf_localidades
```

```python
gdf_localidades["Centroide"]=gdf_localidades.centroid
gdf_localidades
```

```python
gdf_localidades.plot("area", legend=True)
```

```python
axis=gdf_localidades ["geometry"].plot()
gdf_localidades ["Centroide"].plot(ax=axis, color="white")
```

```python
gdf_localidades=gdf_localidades.set_geometry("geometry")
```

```python
gdf_localidades.plot(
    column="LocNombre",
    legend=True,
    edgecolor='black',
    cmap='tab20',
    legend_kwds={"loc":"upper left"},
    figsize=(12,13)
    )
```
