```python
import pandas as pd
```

```python
localidades=pd.read_csv("poblacion.csv" , encoding ='latin1', delimiter=';')
localidades
```

Del Data frame localidades, debe cumplir la condición que la serie año=2024

```python
poblacion_2024=localidades[localidades['Año']==2024]
poblacion_2024
```

```python
grupos=poblacion_2024.groupby("Localidad")
grupos
```

```python
Total_poblacion_localidad=grupos["Población"]. sum()
Total_poblacion_localidad
```

```python
df_poblacion_localidad = Total_poblacion_localidad.reset_index()
df_poblacion_localidad
```

```python
df_poblacion_localidad ['Localidad']=df_poblacion_localidad['Localidad']. str, upper()
df_poblacion_localidad
```
