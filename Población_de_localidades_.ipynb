```python
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx


gdf_localidades = gpd.read_file("localidades.json")


df_poblacion_localidad.rename(columns={"Localidad": "LocNombre"}, inplace=True)


gdf_localidades["poblacion"] = df_poblacion_localidad["Población"]


fig, ax = plt.subplots(figsize=(12, 15))

# Plotear el GeoDataFrame
gdf_localidades.plot(
    ax=ax,
    column="LocNombre",
    legend=True,
    edgecolor='black',
    cmap='Paired',
    legend_kwds={"loc":"upper left", "fontsize": 10 },
    )

# Agregar el mapa base usando contextily
ctx.add_basemap(ax, crs=gdf_localidades.crs)


max_x = gdf_localidades.total_bounds[2]  # Obtener el límite máximo en x
max_y = gdf_localidades.total_bounds[3]  # Obtener el límite máximo en y

leyenda = []
for idx, row in gdf_localidades.iterrows():
    leyenda.append(f"{row['LocNombre']}: {row['poblacion']}")

# Mostrar la lista
for i, txt in enumerate(leyenda):
    ax.text(max_x + 0.05, max_y - 0.02*i, txt, fontsize=8)

# Agregar título y etiquetas de ejes
plt.title('Mapa de Población por Localidad')
plt.xlabel('Longitud')
plt.ylabel('Latitud')


plt.show()
```
